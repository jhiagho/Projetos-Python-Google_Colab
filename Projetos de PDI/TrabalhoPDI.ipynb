# Questão 1.b) função do 2º grau


```python
import math

def equ(a,b,c): #cria a função com tres paramêtros a,b e c
  list = [0,0] #inicia um lista com 2 elementos zerado
  delta = b*b - 4*a*c # calcula o delta
  if (delta < 0): return [] # retorna uma lista vazia. pois não tem como calcular uma raiz de quadrado de um numero negativo
                            # no dominio dos números Reais
  list[0] = (-b + math.sqrt(delta))/2 #aplica a formula.
  list[1] = (-b - math.sqrt(delta))/2
  return list
```

```python
equ(1,1,1)
```

```python
equ(1,-1,-12)
```

```python
equ(1,-4,-5)
```

# Questão 1.c) Fatorial de n

```python
def fatorial(n): #cria uma função com parametro n
  fat = 1 #incia com 1 pois faltoria de 0 e 1
  if (n == 0): return 1 #caso for 0, a resposta será 1
  for i in range(1,n+1): # for que percorre de i = 1 até o valor de n+1
    fat = fat * i #
  return fat
```

```python
def fatorialrec(n): #cria uma função com parametro n
  if (n == 0): return 1 # indica o f(0) = 1
  return n * fatorialrec(n-1) # indica o f(n)= n* f(n-1)
```

```python
fatorial(6)
```

```python
fatorialrec(6)
```

# Questão 1.e) Manipulação de matrizes

```python
import random
```

```python
matriz = [] #inicia a matriz

for i in range(0,5): # percorre a linha indo de 0 a 5
  linha = [] # cria uma lista em cada uma
  for j in range(0,5): # percorre nessa lista criada, o que representa as colunas dessa matriz
    r = random.randrange(-10,10) # armazena em r um número aleátorio indo de -10 a 10
    linha  += [r] #adiciona os valores e essa lista
  matriz += [linha] #acresenta a lista a matriz, o que finaliza a linha
```

```python
matriz
```

```python
soma = 0 #inicia as variáveis
media = 0
cont = 0
cont2 = 0
for i in range(0,5):
  for j in range(0,5):
    if matriz[i][j] < 0: # Difini os valores negativos e conta eles
        cont = cont + 1
        matriz[i][j] = 0  # atribui 0 para esses valores
    if matriz[i][j] > 0: # Difini os valores positivos
        if (matriz[i][j] != 0): #So passará apenas valores positivos difente de zero
            soma = soma + matriz[i][j] #soma esses valores positivos
            cont2 = cont2 + 1 #conta cada um desses numeroes que foi somado
  media = soma / cont2 #calcula médias
```

```python
from PIL import Image

im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
map = im.load() # Cria a matriz de pixel:

# Extra a matriz
w, h = im.size # armazena em w e h as dimensoes da matriz

for i in range(w):
    for j in range(h):
        r, g, b = im.getpixel((i, j)) # Armazena o valor de pixel, no formator (r,g,b)

        # Aplica a formula:
        lena_cinza =  (r + g + b)/3

        # setar um novo valor as pixel, e aplica na imagem original
        map[i, j] = (int(lena_cinza), int(lena_cinza), int(lena_cinza))

im.save("lena_cinza", format="png")
```

```python
im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
map = im.load() # Cria a matriz de pixel:
w, h = im.size

for i in range(w):
    for j in range(h):
        r8, g8, b8 = im.getpixel((i, j))
        #Converte a imagem para 7bits
        sete_bits_r = round(((r8/255))* 127)
        sete_bits_g = round(((g8/255))* 127)
        sete_bits_b = round(((b8/255))* 127)

        map[i, j] = (int(sete_bits_r),int(sete_bits_g),int(sete_bits_b))

        r, g, b = im.getpixel((i, j))
        #Converte de volta para 8bits
        sete_bits_r = round(r / (127) * 255)
        sete_bits_g = round(g / (127) * 255)
        sete_bits_b = round(b / (127) * 255)

        map[i, j] = (int(sete_bits_r),int(sete_bits_g),int(sete_bits_b))
im.save("lena_7con", format="png")
im = Image.open("lena_cor.bmp") # importa a imagem do diretorio:
im2 = Image.open("lena_7con")
fig, axs = plt.subplots(ncols=2, figsize=(10, 10))
axs[0].imshow(im, cmap=plt.cm.gray)
axs[0].set_title("Normal 8bits")
axs[1].imshow(im2, cmap=plt.cm.gray)
axs[1].set_title("Quantizada em 7bits")
```
